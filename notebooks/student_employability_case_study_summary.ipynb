# Student Employability Prediction Using Machine Learning

This notebook summarizes a portfolio case study for predicting student placement readiness from academic and skill-related indicators.

## Objective

The goal is to identify the strongest factors related to student placement and compare machine learning models that classify students as placed or not placed.

## Dataset

The project uses a public Campus Recruitment CSV dataset. It includes CGPA, internships, projects, certifications, aptitude score, soft skills, extracurricular activities, placement training, academic marks, and placement status.

```python
import pandas as pd

summary = pd.DataFrame({
    'metric': ['records', 'placement_rate', 'best_model', 'validation_accuracy', 'validation_f1', 'validation_roc_auc'],
    'value': [9957, 0.4173, 'Logistic Regression', 0.8000, 0.7637, 0.8758]
})
summary
```

```python
model_performance = pd.DataFrame({
    'model': ['Logistic Regression', 'Random Forest', 'Gradient Boosting'],
    'validation_accuracy': [0.8000, 0.7984, 0.8016],
    'validation_f1': [0.7637, 0.7559, 0.7606],
    'validation_roc_auc': [0.8758, 0.8732, 0.8714]
})
model_performance
```

## Executive Results

- Logistic Regression was selected as the best validation model by ROC AUC.
- The model reached 80.00% validation accuracy.
- Validation ROC AUC reached 87.58%.
